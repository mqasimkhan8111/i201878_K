# Projection 

```python
def visualize(vector, v2, proj, proj2):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)

    ax.grid(alpha=0.8)
    ax.set(xlim=(-25, 25), ylim=(-25, 25))


    # Plot the orthogonal vectors
    ax.annotate('', xy=vector, xytext=(0, 0), arrowprops=dict(facecolor='r'))
    ax.annotate('', xy=v2, xytext=(0, 0), arrowprops=dict(facecolor='g'))
    ax.annotate('', xy=proj, xytext=(0, 0), arrowprops=dict(facecolor='b'))
    ax.annotate('', xy=proj2, xytext=(0, 0), arrowprops=dict(facecolor='y'))
    plt.show()
```

## Projection of v1 on v2 

```python
v1=vector
mag_v2= (np.sqrt(np.sum(v2 * v2)))
mag_v1= (np.sqrt(np.sum(v1 * v1)))
proj = np.dot(v1,v2)*v2/mag_v2
```

```python
proj
```

## Projection of v2 on v1 

```python
mag_v2= (np.sqrt(np.sum(v2 * v2)))
proj2 = np.dot(v2,v1)*v1/mag_v1
```

```python
proj2
```

```python
visualize(v1, v2, proj, proj2)
```
