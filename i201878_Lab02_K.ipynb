# Parent Class

```python
class University:
    _campuses=5
    def __init__(self,uname,location):
        self.uname=uname
        self.location=location
    def display(self):
        #print("Name is ",self.name + "Location is ",self.location)
        print(f"Name is {self.uname} and location is {self.location}")
        #print(self._campuses)
```

```python
obj1 = University("FAST", "Islamabad")
obj1.campuses=6
print(obj1.campuses)
obj1.display()
```

# Child Class

```python
class Department(University):
    def __init__(self,name,programs, u_name,u_location):
        super().__init__(u_name,u_location)
        self.name=name
        self.programs=programs
    def display(self):
        super().display()
        print(f"The name is {self.name} and programs are {self.programs}")
```

```python
obj1 = Department("AI", "5", "FAST", "Islamabad")
obj1.display()
```

```python
def func1(strings):
    count = 0
    for i in range(0, len(strings)):
        print(strings[i])
        if(len(strings)>=2) and len(strings[i])>1:
            if(strings[i][0] == strings[i][len(strings[i]) - 1]):
                count+=1
    return count
```

```python
strings = ["Ali is not well", "Qasim's name is misaQ", "Qq", "QQ", "A"]
func1(strings)
```

```python
def func2(strings):
    count = []
    decrement = 0
    temp_l=[]
    for i in range(0, len(strings)):
        if(strings[i][0]=='x'):        #Only checking small 'x' condition here...
            temp_l.append(strings[i])
            count.append(i)
    for i in count:
        strings.pop(i-decrement)
        decrement+=1

    strings.sort()
    sorted_list=[]
    sorted_list = temp_l + strings
    return sorted_list
```

```python
strings = ["my name is khan", "xhy", "may Allah bless m", "xavier"]
func2(strings)
```

```python
def func3(tuples):
    count = len(tuples) 
    for i in range(0, count):     
        for j in range(count-1, i+0, -1): 
            if (tuples[j][-1] < tuples[j - 1][-1]): 
                temp = tuples[j] 
                tuples[j]= tuples[j - 1] 
                tuples[j - 1]= temp 
    return tuples 
```

```python
tuples =[(1, 2), (1, 3), (8, 1), (5, 0)]
func3(tuples)
```

```python
def func4(l1):
    i = len(l1) -1
    while(i>0):
        if(l1[i]==l1[i-1]):
            l1.pop(i)
        i-=1
    return l1
```

```python
l1=[1,2,2,2,2,3,2,5,5,5,7,6,5]
func4(l1)
```

```python
def func5(l1, l2):
    merged_list=[]
    merged_list = l2 + l1
    merged_list.sort()
    return merged_list
```
