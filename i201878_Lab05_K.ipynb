```python
balls= ['w', 'r', 'b']
combs=[]
for i in balls:
    for j in balls:
        for k in balls:
            combs.append(i+j+k)
combs
```

```python
w=6
r=4
b=5
prob_w = w/15
prob_r = r/15
prob_b = b/15
```

```python
def combinations(n,r):
    n_fact=1
    n=n
    r=r

    for i in range(1,n+1):
        n_fact*=i

    nsubr = n - r
    nsubr_fact = 1
    for i in range(1,nsubr + 1):
        nsubr_fact*=i
    r_fact=1
    for i in range(1, r+1):
        r_fact*=i
    return (n_fact)/(nsubr_fact * r_fact)
```

```python
combs_prob = []
prob=1
n_w=6
n_r=4
n_b=5
r_w=0
r_r=0
r_b=0

for i in combs:
    prob_w = 6/15
    prob_r = 4/15
    prob_b = 5/15
    
    for j in i:
        if(j=='w'):
            r_w+=1
        if(j=='r'):
            r_r+=1
        if(j=='b'):
            r_b+=1
    
    a=combinations(n_w,r_w)
    b=combinations(n_r,r_r)
    c=combinations(n_b,r_b)
    d=combinations(15,3)
    
    r_w=0
    r_r=0
    r_b=0
    
    combs_prob.append((a*b*c)/d)
    prob = 1
index = combs_prob.index(max(combs_prob))
```

```python
print("Optimal Color Scheme : ", combs[index])
print("Optimal Color Scheme Probability : ", max(combs_prob))
```

```python
def calculate_fx(x, covList, Means_v):
    temp=((2*3.142)**4)
    temp1=np.linalg.det(covList)
    temp2 = temp * temp1
    temp2= temp2**(1/2)
    temp5 = (x - Means_v).T
    temp3= (-0.5) * temp5
    temp4 = np.dot(np.linalg.inv(covList),  (x - Means_v))
    temp6 = (np.dot(temp3, temp4))
    temp5 = 2.71828**temp6

    fx=temp2 * temp5
    return fx
```
